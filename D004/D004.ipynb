# Sample

- Rotating
- Scaling
- Shifting

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

img_path = '../lena.png'
img = cv2.imread(img_path)
```

## Rotation

```python
# 垂直翻轉 (vertical)
img_vflip = img[::-1, :, ::-1]
plt.imshow(img_vflip)
```

## Scaling

down size the img first to see how it works when up sizing imgages.  
`csv.resize(img, [resize_shape, fx, fy](either choose one), [interpolation])`

```python
img_small = cv2.resize(img, None, fx = 0.2, fy = 0.2)
plt.imshow(img_small[:, :, ::-1])
plt.axis('off')
plt.show()
```

```python
img_cubic = cv2.resize(img_small, None, fx = 5, fy = 5, interpolation = cv2.INTER_CUBIC)
img_area = cv2.resize(img_small, None, fx = 5, fy = 5, interpolation = cv2.INTER_AREA)
img_lanczos = cv2.resize(img_small, None, fx = 5, fy = 5, interpolation = cv2.INTER_LANCZOS4)
```

```python
img_scale = np.hstack((img_cubic, img_area, img_lanczos))
plt.imshow(img_scale[:, :, ::-1])
plt.axis('off')
plt.show()
```

```python
while True:
    cv2.imshow('zoom image', img_scale)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

## Shifting

- use `warpAffine()`  

- x += 100; y += 50

```python
M = np.array([[1, 0, 100],
              [0, 1, 50]], dtype=np.float32)
img_shift = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
```

```python
while True:
    cv2.imshow('shift image', img_shift)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

# HW

```python
img_vflip = img[::-1, :, ::-1]
img_hflip = img[:, ::-1, ::-1]
img_hvflip = img[::-1, ::-1, ::-1]
img_flip = np.vstack((np.hstack((img[:, :, ::-1], img_hflip)), np.hstack((img_vflip, img_hvflip))))

plt.imshow(img_flip)
plt.axis('off')
plt.show()
```
