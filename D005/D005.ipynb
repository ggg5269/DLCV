# Sample

plotting on the image:  
- equalize the brightness
- horizontally flip & scaling to 50%
- outline Lena

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

img_path = '../lena.png'
img = cv2.imread(img_path)
```

## Equalize the Brightness

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsv[..., -1] = cv2.equalizeHist(img_hsv[..., -1])
img_eq = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)
while True:
    cv2.imshow('equalized', img_eq)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

```python
plt.imshow(img_eq[..., ::-1])
plt.axis('off')
plt.show()
```

## Flip

```python
img_zoom = cv2.resize(img_eq, None, fx = 0.5, fy = 0.5)[:, ::-1, :]
```

```python
plt.imshow(img_zoom[..., ::-1])
plt.axis('off')
plt.show()
```

## Graph a Rectangle

```python
img_rec = cv2.rectangle(img_zoom, (30, 20), (210, 255), (255, 255, 255), 3)
img_text = cv2.putText(img_rec, 'Lena', (80, 250), 6, 1, (255, 255, 255), 2)
```

```python
while True:
    cv2.imshow('image', img_text)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```
